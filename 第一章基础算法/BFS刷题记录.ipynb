```python
# %% 广度优先遍历模版1
# 模板1,不带hash表，每个节点可以访问多次，遍历树没有问题，遍历图就会陷入循环
def BFS1(root, target):
    # 记录需要的步数(遍历的层数)
    step = 0
    # 这一层的节点列表，注意第一层要将根节点放进去
    this_layer = [root]
    # 当这一层的节点列表不为空，就要一直遍历
    while this_layer != []:
        # 下一层的节点列表
        next_layer = []
        # 遍历这一层的节点
        for node in this_layer:
            if node.value == target:
                return step  # 找到了就返回
            # 同时要把每个节点的下一个节点加入到下一层节点列表
            next_layer.append(node.next)
        # 遍历完这一层的节点，step就加一
        step += 1
        this_layer = next_layer


# %%
# 模板2，带hash表，保证每个节点只访问一次，遍历图没有问题
def BFS2(root, target):
    # 只需要在BFS1的基础上将遍历过的节点加入hash表就好了
    hash_table = set()
    # 记录遍历的层次
    step = 0
    # 这一层的节点列表，注意第一层要将根节点放进去
    this_layer = [root]
    # 当这一层的节点列表不为空，就要一直遍历
    while this_layer != []:
        # 下一层的节点列表
        next_layer = []
        # 遍历这一层的节点
        for node in this_layer:
            # 没遍历过才需要遍历
            if node not in hash_table:
                if node.value == target:
                    return step  # 找到了就返回
                # 被遍历了就加入hash_table
                hash_table.add(node)
            # 同时要把每个节点的下一个节点加入到下一层节点列表
            next_layer.append(node.next)
        # 遍历完这一层的节点，step就加一
        step += 1
        this_layer = next_layer
```

#### 1、岛屿数量
给你一个由 '1'（陆地）和 '0'（水）组成的的二维网格，请你计算网格中岛屿的数量。

岛屿总是被水包围，并且每座岛屿只能由水平方向和/或竖直方向上相邻的陆地连接形成。

此外，你可以假设该网格的四条边均被水包围。

例如 :  
输入：grid = {  
  &nbsp;["1","1","1","1","0"],  
  &nbsp;["1","1","0","1","0"],  
  &nbsp;["1","1","0","0","0"],  
  &nbsp;&nbsp;&nbsp;&nbsp;["0","0","0","0","0"]  
}  
输出：1

链接：https://leetcode.cn/leetbook/read/queue-stack/kbcqv/  
来源：力扣（LeetCode）  

```python
# 广度优先遍历 写法1
class Solution:
    def numIslands(self, grid: list[list[str]]) -> int:
        # 获取网格的行数
        num_row = len(grid)
        # 行数为0 直接返回即可
        if num_row == 0:
            return 
        # 获取网格的列数
        num_col = len(grid[0])
        num_islands = 0  # 岛屿的数量
        queue = list()  # 队列

        # 遍历二维网格的每一个位置
        for i in range(num_row):
            for j in range(num_col):
                # 如果未访问过
                if grid[i][j] == '1':
                    num_islands += 1    # 岛屿数目+1
                    grid[i][j] = '0'   # 标记为未访问
                    queue.append((i,j))  # 当前位置访问后 加入队列借以访问周边元素

                    # 访问队列中元素的周边元素情况
                    while queue :
                        # 获取位置
                        row , col = queue.pop(0)
                        # 遍历上下左右
                        for x,y in [(row , col+1),(row+1 , col),(row , col-1),(row-1 , col)]:
                            # 如果在范围内且也没有被访问过 直接标记为已访问过
                            # 因为题意要求 记录上下左右与当前位置不同时 is_lands++
                            if 0<=x<num_row and 0<=y<num_col and grid[x][y]== '1':
                                # 同理 加入队列以访问周边元素的情况
                                queue.append((x,y))
                                grid[x][y] = '0'    # 标记为已访问
        # 整个大循环后 返回岛屿的数目
        return num_islands
```

```python
# 广度优先遍历 写法2
from collections import deque

class Solution:
    def numIslands(self, grid: list[list[str]]) -> int:
        # 获取二维网格的行数
        num_row = len(grid)
        if num_row == 0:
            return 0  # 如果网格为空，直接返回0

        # 获取二维网格的列数
        num_col = len(grid[0])
        num_islands = 0  # 初始化岛屿数量为0

        # 遍历二维网格
        for i in range(num_row):
            for j in range(num_col):
                # 如果当前位置是陆地 '1'(这里满足条件的 1 也可以看做未访问)
                if grid[i][j] == '1':
                    num_islands += 1  # 岛屿数量加1
                    grid[i][j] = '0'  # 将当前位置标记为已访问
                    queue = deque([(i, j)])  # 初始化双端队列，并将当前位置加入队列
                    # 双端队列存在的意义是为了搜索队列中元素的周围元素是否满足条件
                    # 采用队列是为了规范搜索顺序

                    # 进行广度优先搜索
                    while queue:
                        row, col = queue.popleft()  # 出队列一个位置
                        directions = [(row, col + 1), (row + 1, col), (row, col - 1), (row - 1, col)]

                        # 遍历当前位置的上、下、左、右四个相邻位置
                        for x, y in directions:
                            # 判断相邻位置是否在合法范围内，并且是否是陆地 '1'
                            # 排除周围元素中不满足条件的元素
                            if 0 <= x < num_row and 0 <= y < num_col and grid[x][y] == '1':
                                queue.append((x, y))  # 将相邻位置加入队列
                                grid[x][y] = '0'  # 标记相邻位置为已访问

        # 返回最终岛屿数量
        return num_islands
```
